# Pandas Dataframes

```python
import pandas as pd
```

## Cria um dicionario com autores, preços e títulos

```python
dictionary = {
    'Autores' : [
        'Rick Riordan',
        'J. R. R. Tolkien',
        'Rick Riordan',
        'Machado de Assis'
    ],
    'Titulos' : [
        'O Ladrão de Raios',
        'A Sociedade do Anel',
        'Mar de Monstros',
        'Memórias Póstumas de Brás Cubas'
    ],
    'Preços' : [
        41.2,
        35.7,
        39.5,
        40.5
    ]
}
```

## DataFrame

```python
df = pd.DataFrame(dictionary)
df
```

```python
df['Autores']
```

```python
df['Preços'].mean()
```

```python
mascara = df['Autores'] == 'Rick Riordan'
```

```python
print(mascara)
```

```python
df[mascara]
```

```python
new_df = pd.DataFrame({'Autores' : ['Rick Riordan'], 'Titulos' : ['A Maldiça do Titã'], 'Preços' : [40.2]})
df = pd.concat([df, new_df], ignore_index = True)
df
```

```python
mascara_correcao = df['Titulos'] == 'A Maldiça do Titã'
```

```python
df.loc[mascara_correcao, 'Titulos'] = 'A Maldição do Titã'
df
```
